### Gradient Descent Lemonade Stand Example

The model of the profit of a lemonade stand is given by the equation:

$$
\text{profit} = -4 x^2 + 40x
$$

where $x$ is the price of a cup of lemonade.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
profit = -4*x**2 + 40*x

# create a function that returns the profit
def profit_function(x):
    return -4*x**2 + 40*x


plt.plot(x, profit)
```

The gradient of the profit function is given by:

$$
\frac{d\text{profit}}{dx} = -8x + 40
$$

```python
gradient = -8*x + 40

# create a function that returns the gradient
def gradient_function(x):
    return -8*x + 40

plt.plot(x, gradient)
```

```python
# plot the function
# plt.plot(x, profit)

# plot the gradient descent in the same plot, 100 steps
x0 = 1
for i in range(100):
    x0 = x0 - 0.01*gradient_function(x0)
    y = profit_function(x0)
    plt.scatter(x0, y, color='red')

plt.show()

```
